```python
import pandas as pd #Importing pandas library

#Reading the csv file
dfProd = pd.read_csv('producao.csv', sep=';')

#Printing the first 5 rows of the csv file
dfProd.head() 
```

Todo tratamento feito na base de modelo, por isso, vamos actualizar a base de dados com a base de produção
Já não se tem as colunas seguro_incendio,taxa_serviço e total nessa nova base, só falta tratar dos  bairros

```python
#importing the file regions.csv
regiao = pd.read_excel('bairros.xlsx')

print(regiao.head())

```

```python
# Merge the two files dfProd and regiao
dfProd = pd.merge(dfProd, regiao, how='left', left_on='bairro', right_on='Bairro')

#Printing the first 5 rows of the csv file
dfProd.head()

#dfProd = pd.merge(dfProd, regiao, on='Bairro', how='left')
```

```python
#Deleting the column 'Bairro' and 'bairro'
dfProd = dfProd.drop(columns=['Bairro', 'bairro'])
```

```python
#Putting One Hot Encoding in the column 'Regiao'
dfProd = pd.get_dummies(dfProd, columns=['Região'])
```

```python
dfProd.head()
```

```python
#Deliting the column 'Região'
dfProd = dfProd.drop(columns=['Região'])
```

```python
#Changing values of the column 'Região_Centro	Região_Leste	Região_Norte	Região_Oeste	Região_Sul' where is true put 1 and false put 0
dfProd['Região_Centro'] = dfProd['Região_Centro'].astype(int)
dfProd['Região_Leste'] = dfProd['Região_Leste'].astype(int)
dfProd['Região_Norte'] = dfProd['Região_Norte'].astype(int)
dfProd['Região_Oeste'] = dfProd['Região_Oeste'].astype(int)
dfProd['Região_Sul'] = dfProd['Região_Sul'].astype(int)
dfProd.head()
```

```python
# Importing load from joblib
from joblib import load

# Loading the model
model = load(r'C:\Users\Elvis\Documents\GitHub\Modelo em produção\Market-Value-Predictor-Airbnb\regRandomForest3.joblib')
```

```python
#Deliting the column 'aluguel' and putting the values in x
x = dfProd.drop(columns=['aluguel'])
#Putting the values of the column 'aluguel' in y
y = dfProd['aluguel']


```

```python
# Renaming the columns
x = x.rename(columns={
    'Região_Centro': 'Centro',
    'Região_Leste': 'Leste',
    'Região_Norte': 'Norte',
    'Região_Oeste': 'Oeste',
    'Região_Sul': 'Sul'
})

# Making the prediction with the new data
y_pred = model.predict(x)

# Printing the prediction
print(y_pred)
```

```python
# Adding the prediction to the dataframe
dfProd['aluguel'] = y_pred

# Printing the dataframe
dfProd.head()
```
