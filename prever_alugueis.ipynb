```python
#importing the file regions.csv
regiao = pd.read_excel('regions.csv')

#Merge the two files
dfProd = pd.merge(
    dfProd,
    regiao, 
    how='left',
    left_on='bairro',
    right_on='Bairro')
    
    #Deleting the column 'Bairro' and 'bairro'
dfProd = dfProd.drop(columns=['Bairro', 'bairro'], axis=1)




```
